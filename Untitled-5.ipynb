```python
# Import Required Libraries
import pandas as pd

# Load CSV File with specified encoding
df = pd.read_csv(r'c:\Users\User\Downloads\opendata_ตัวอย่างข้อมูล_เงินฝาก.csv', encoding='cp874')

# Display First Few Rows
df.head()
```

```python
# แปลงข้อมูลในคอลัมน์ province และ sex ให้เป็นภาษาอังกฤษ
df['province'] = df['province'].replace({'กรุงเทพมหานคร': 'Bangkok', 'สมุทรสาคร': 'Samut Sakhon', 'บุรีรัมย์': 'Buriram'})
df['sex'] = df['sex'].replace({'ชาย': 'Male', 'หญิง': 'Female'})

# แสดงข้อมูลที่ถูกแปลงแล้ว
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt

# Fit the KMeans model
# Encode categorical features
label_encoder_sex = LabelEncoder()
label_encoder_province = LabelEncoder()

df['sex_encoded'] = label_encoder_sex.fit_transform(df['sex'])
df['province_encoded'] = label_encoder_province.fit_transform(df['province'])

# Select features for clustering
features = ['sex_encoded', 'province_encoded', 'age', 'DEP_TOTAL']
encoded_features = df[features]

# Fit the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(encoded_features)

# Add the cluster labels to the dataframe
df['cluster'] = kmeans.labels_

# Display the dataframe with the cluster labels
df.head()

# Plot the clusters
plt.scatter(df['age'], df['DEP_TOTAL'], c=df['cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('DEP_TOTAL')
plt.title('Clusters of Customers')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the clusters
scatter = ax.scatter(df['age'], df['DEP_TOTAL'], df['province_encoded'], c=df['cluster'], cmap='viridis')

# Set labels
ax.set_xlabel('Age')
ax.set_ylabel('DEP_TOTAL')
ax.set_zlabel('Province Encoded')
ax.set_title('3D Clusters of Customers')

# Add legend
legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend1)

plt.show()
```

```python
# Create a 4D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the clusters with color representing DEP_TOTAL
scatter = ax.scatter(df['age'], df['province_encoded'], df['sex_encoded'], c=df['DEP_TOTAL'], cmap='viridis')

# Set labels
ax.set_xlabel('Age')
ax.set_ylabel('Province Encoded')
ax.set_zlabel('Sex Encoded')
ax.set_title('4D Clusters of Customers')

# Add color bar
cbar = fig.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)
cbar.set_label('DEP_TOTAL')

plt.show()
```

```python
# Filter out rows where age is greater than 100
df_filtered = df[df['age'] <= 100]

# Create a 4D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the clusters with color representing DEP_TOTAL
scatter = ax.scatter(df_filtered['age'], df_filtered['province_encoded'], df_filtered['sex_encoded'], c=df_filtered['DEP_TOTAL'], cmap='viridis')

# Set labels
ax.set_xlabel('Age')
ax.set_ylabel('Province Encoded')
ax.set_zlabel('Sex Encoded')
ax.set_title('4D Clusters of Customers (Age <= 100)')

# Add color bar
cbar = fig.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)
cbar.set_label('DEP_TOTAL')

plt.show()
```

```python
# Create a scatter plot with 'sex' and 'province' as axes
plt.figure(figsize=(10, 6))
plt.scatter(df['sex'], df['province'], c=df['cluster'], cmap='viridis')
plt.xlabel('Sex')
plt.ylabel('Province')
plt.title('Clusters of Customers by Sex and Province')
plt.show()
```
